```python
import pretty_midi

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd
import os
```

```python
# 读取metadata的CSV文件
metadata_file = 'maestro-v2.0.0/maestro-v2.0.0.csv'
metadata_df = pd.read_csv(metadata_file) 

# 遍历每一行，加载并处理midi和mp3文件
for index, row in metadata_df.iterrows():
    if index<41:
        continue
    
    # 读取midi和mp3文件名
    midi_filename = row['midi_filename']
    audio_filename = row['audio_filename']
    split = row['split']  # 获取 'split' 列的值
    # 将音频文件名的后缀修改为 mp3
    audio_filename = audio_filename.replace('.wav', '.mp3')

    print("MIDI文件名:", midi_filename)
    print("音频文件名:", audio_filename)

    # 拼接文件路径
    midi_path = os.path.join('maestro-v2.0.0/', midi_filename.strip()) # 去除字符串两端的空格
    audio_path = os.path.join('maestro-v2.0.0/', audio_filename.strip()) # 去除字符串两端的空格

    print("MIDI路径:", midi_path)
    print("音频路径:", audio_path)

    # Load MIDI file into PrettyMIDI object
    midi_data = pretty_midi.PrettyMIDI(midi_path)
    total_time = midi_data.get_end_time()
    print("MIDI duration:", total_time, "seconds")
    # Get piano-roll of midi (convert to matrix)
    midi_matrix = midi_data.get_piano_roll(fs=50)
    # print("Piano Roll: ", midi_matrix.shape,"\n", midi_matrix)

    # Load the MP3 file
    audio_data, sr = librosa.load(audio_path, duration=midi_data.get_end_time())  # 指定长度为midi文件的长度 default sr=22050
    print("MP3 Duration: ", librosa.get_duration(y=audio_data), "seconds")
    print("Sample Rate: ", sr, "Hz")
    mel_spectrogram = librosa.feature.melspectrogram(y=audio_data, sr=sr, n_fft=2048, hop_length=441, n_mels=128)
    audio_matrix = librosa.power_to_db(mel_spectrogram, ref=np.max)
    # audio_matrix = librosa.stft(audio_data, hop_length=245)
    # audio_matrix = np.abs(audio_matrix)  # get amplitude 取幅度
    # print("Mel Audio-Matrix:",audio_matrix.shape,"\nMAX and MIN",np.max(audio_matrix),np.min(audio_matrix)) #,"\n",audio_matrix)

    midi_duration = midi_data.get_end_time()
    audio_duration = librosa.get_duration(y=audio_data, sr=sr)

    # print("MIDI duration:",midi_duration)
    # print("MP3 Duration: ",audio_duration)

    print("MIDI Shape: ", midi_matrix.shape)
    print("MP3 Shape: ", audio_matrix.shape)
    
    # 定义片段长度（3 秒对应的帧数） 
    segment_length = 3 # 3 second
    fps=50 # frame per second
    columns_per_segment = segment_length * fps  

    # 计算可以切割出多少个 3 秒片段（取整，不足舍去）
    num_segments = midi_matrix.shape[1] // columns_per_segment
    print(f"There are {num_segments} segments in file {index+1}\n")
    
    # 存储切割后的片段 保存路径 # 根据 'split' 的值确定保存路径
    if split == 'train':
        save_folder = "./autodl-tmp/datasetF/train"
    elif split == 'validation':
        save_folder = "./autodl-tmp/datasetF/validation"
    elif split == 'test':
        save_folder = "./autodl-tmp/datasetF/test"
    else:
        # 如果 'split' 列的值不是 'train', 'validation' 或 'test', 可以根据需要进行处理
        save_folder = "./autodl-tmp/datasetF/other"

    # 确保保存路径存在
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    # 遍历 MIDI 和 MP3 的矩阵，将它们切片成对应的 3 秒片段
    for i in range(num_segments):
        start_col = i * columns_per_segment
        end_col = start_col + columns_per_segment
        
        # 切片 MIDI 矩阵
        midi_segment = midi_matrix[:, start_col:end_col]
        
        # 切片 MP3 矩阵
        audio_segment = audio_matrix[:, start_col:end_col]
        
        # 生成文件名
        segment_number = i + 1
        midi_seg_filename = f"midi-{index+1:04d}_seg-{segment_number:04d}.npy"  # MIDI 文件名
        audio_seg_filename =f"audio-{index+1:04d}_seg-{segment_number:04d}.npy"  # MP3 文件名
        
        # 确定保存路径
        midi_save_path = os.path.join(save_folder, midi_seg_filename)
        audio_save_path = os.path.join(save_folder, audio_seg_filename)
        
        # 保存片段矩阵
        np.save(midi_save_path, midi_segment)
        np.save(audio_save_path, audio_segment)

    print("Line", index+1, " is done.\n")
```
