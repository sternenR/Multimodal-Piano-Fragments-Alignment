```python
import numpy as np
import os
import matplotlib.pyplot as plt

import torch                     # for all things PyTorch
import torch.nn as nn            # for torch.nn.Module, the parent object for PyTorch models
import torch.nn.functional as F  # for the activation function
import torchvision.models as models
import torch.optim as optim

from torch.utils.data import Dataset,DataLoader
from torchvision import datasets, transforms
```

```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)
print(f"Using {device} device")
```

```python
# DataSet & DataLoader
class PairedDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.file_names = os.listdir(root_dir)
        self.paired_samples = []

        npy_files = [f for f in self.file_names if f.endswith('.npy')]
        # 找到所有匹配的样本对（正样本）
        for midi_name in npy_files:
            if 'midi' in midi_name:
                audio_name = midi_name.replace('midi', 'audio')
                if audio_name in npy_files:
                    self.paired_samples.append((midi_name, audio_name))
                    
        print(len(self.paired_samples)," == " , len(npy_files)//2) ##

    def __len__(self):
        return len(self.paired_samples)

    def __getitem__(self, idx):
        # 获取正样本（匹配的样本对）
        midi_name, audio_name = self.paired_samples[idx]

        # 加载正样本数据
        midi_path = os.path.join(self.root_dir, midi_name)
        audio_path = os.path.join(self.root_dir, audio_name)
        
        midi_np = np.load(midi_path)
        audio_np = np.load(audio_path)
        
        # 归一化音频数据
        audio_min, audio_max = np.min(audio_np), np.max(audio_np)
        if audio_max != audio_min:  # 避免除以零
            audio_np = (audio_np - audio_min) / (audio_max - audio_min)
        else:
            audio_np = np.zeros_like(audio_np)

        # 归一化MIDI数据
        midi_min, midi_max = np.min(midi_np), np.max(midi_np)
        if midi_max != midi_min:  # 避免除以零
            midi_np = (midi_np - midi_min) / (midi_max - midi_min)
        else:
            midi_np = np.zeros_like(midi_np)

        # 将 NumPy 数组转换为 PyTorch 张量 
        midi_data = torch.tensor(midi_np, dtype=torch.float32).unsqueeze(0)
        audio_data = torch.tensor(audio_np, dtype=torch.float32).unsqueeze(0)

        # 如果需要，应用变换
        if self.transform:
            midi_data = self.transform(midi_data)
            audio_data = self.transform(audio_data)

        # 返回正样本和标签
        return midi_data, audio_data
```

```python
class ResNet_Encoder(nn.Module):
    def __init__(self, input_channels):
        super(ResNet_Encoder, self).__init__()
        # Load the pretrained ResNet model without fully connected layers
        self.resnet = models.resnet18(pretrained=True)
        # Modify the initial convolutional layer to accept input_channels
        self.resnet.conv1 = nn.Conv2d(input_channels, 64, kernel_size=7, stride=2, padding=3, bias=False)
        self.resnet.fc = nn.Identity()  # Remove the final fully connected layer
        
        # # Set requires_grad to False to freeze the parameters
        # for param in self.resnet.parameters():
        #     param.requires_grad = False

    def forward(self, x):
        # Forward pass through ResNet-18
        features = self.resnet(x)
        # Apply L2 normalization
        features = F.normalize(features, p=2, dim=1)
        # Flatten the features
        features = features.view(features.size(0), -1)
        return features
```

```python
class ProjectionHead(nn.Module):
    def __init__(self, input_dim=512, hidden_dim=256, output_dim=128):
        super(ProjectionHead, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)  
        self.fc3 = nn.Linear(hidden_dim, output_dim)
        # self.bn = nn.BatchNorm1d(output_dim)  # Batch Normalization layer

    def forward(self, x):           
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x)) 
        x = self.fc3(x)
        
        # x = self.bn(x)  # Apply Batch Normalization
        return x
```

```python
class MusicCLIP(nn.Module):
    def __init__(self):
        super(MusicCLIP, self).__init__()
        self.midi_encoder = ResNet_Encoder(input_channels=1)
        self.audio_encoder = ResNet_Encoder(input_channels=1)
        # proj = ProjectionHead(input_dim, hidden_dim, output_dim)
        self.midi_projection_head = ProjectionHead()
        self.audio_projection_head = ProjectionHead()
        
        # self.scale = nn.Parameter(torch.tensor(1.0))  # Learnable scale parameter initialized to 1.0


    def forward(self, midi_data, audio_data):
        midi_features = self.midi_encoder(midi_data)
        audio_features = self.audio_encoder(audio_data)
        midi_features = self.midi_projection_head(midi_features)
        audio_features = self.audio_projection_head(audio_features)
        
        return midi_features, audio_features
        
#         # normalized features 计算范数，也就是先标准化
#         midi_features = midi_features / midi_features.norm(dim=1, keepdim=True)
#         audio_features = audio_features / audio_features.norm(dim=1, keepdim=True)
        
#         # Calculate cosine similarity matrix 然后再计算才是余弦相似度 and apply scale
#         similarity_matrix = midi_features @ audio_features.t()
#         similarity_matrix *= self.scale       
        
#         return similarity_matrix, midi_features, audio_features
```

```python
def calculate_similarity_matrix(midi_features, audio_features):
    # Normalize features
    midi_features = midi_features / midi_features.norm(dim=1, keepdim=True)
    audio_features = audio_features / audio_features.norm(dim=1, keepdim=True)
    
    # Calculate cosine similarity matrix
    similarity_matrix = midi_features @ audio_features.t()
    
    return similarity_matrix


class Symmetric_CrossEntropyLoss(nn.Module):
    def __init__(self):
        super(Symmetric_CrossEntropyLoss, self).__init__()
        self.logit_scale = nn.Parameter(torch.ones([]) * np.log(1/0.07))  # Initialize to equivalent of 0.07 as CLIP advised
        # self.logit_scale = nn.Parameter(torch.ones([]))  # Initialize the logit scale parameter, nn.Parameter makes it learnable.

    def forward(self, midi_features, audio_features):
        # Calculate similarity matrix
        similarity_matrix = calculate_similarity_matrix(midi_features, audio_features)
        
        batch_size = similarity_matrix.size(0)
        
        # Apply the learnable temperature (logit scale)
        logit_scale = self.logit_scale.exp()
        similarity_matrix = similarity_matrix * logit_scale
        
        # Target labels: 0, 1, 2, ..., batch_size-1
        labels = torch.arange(batch_size).to(similarity_matrix.device)
        
        # Cross-entropy loss for midi-to-audio
        loss_midi_to_audio = F.cross_entropy(similarity_matrix, labels)
        
        # Cross-entropy loss for audio-to-midi
        loss_audio_to_midi = F.cross_entropy(similarity_matrix.t(), labels)
        
        # Total loss
        loss = (loss_midi_to_audio + loss_audio_to_midi) / 2
        return loss
```

```python
batch_size = 256

# 创建训练数据集实例和数据加载器
train_dataset = PairedDataset(root_dir='./autodl-tmp/datasetF/train')
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# 定义验证数据集和数据加载器
val_dataset = PairedDataset(root_dir='./autodl-tmp/datasetF/validation')
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
```

```python
lr = 1e-4
epochs = 80

model = MusicCLIP()
loss_fn = Symmetric_CrossEntropyLoss() 

model.to(device)

optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)  # 使用 Adam 优化器

# from torch.optim.lr_scheduler import MultiStepLR
# scheduler = MultiStepLR(optimizer, milestones=[1,3,5,8], gamma=0.5) #,10,15,20,23,28

# from torch.optim.lr_scheduler import ReduceLROnPlateau
# scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5, verbose=True)
```

```python
def calculate_accuracy(midi_features, audio_features):
    # Calculate similarity matrix
    similarity_matrix = calculate_similarity_matrix(midi_features, audio_features)
    
    # Get batch size
    batch_size = similarity_matrix.size(0)
    
    # Target labels: 0, 1, 2, ..., batch_size-1
    labels = torch.arange(batch_size).to(similarity_matrix.device)
    
    # Find the index of the maximum value in each row (midi-to-audio)
    max_indices_midi_to_audio = similarity_matrix.argmax(dim=1)
    # Find the index of the maximum value in each column (audio-to-midi)
    max_indices_audio_to_midi = similarity_matrix.argmax(dim=0)
    
    # Calculate the number of correct matches for midi-to-audio
    correct_midi_to_audio = (max_indices_midi_to_audio == labels).sum().item()
    # Calculate the number of correct matches for audio-to-midi
    correct_audio_to_midi = (max_indices_audio_to_midi == labels).sum().item()
    
    # Total accuracy
    total_correct = correct_midi_to_audio + correct_audio_to_midi
    total_samples = 2 * batch_size  # Because we are counting both midi-to-audio and audio-to-midi
    
    accuracy = total_correct / total_samples
    return accuracy

def calculate_similarity_matrix(midi_features, audio_features):
    # Normalize features
    midi_features = midi_features / midi_features.norm(dim=1, keepdim=True)
    audio_features = audio_features / audio_features.norm(dim=1, keepdim=True)
    
    # Calculate cosine similarity matrix
    similarity_matrix = midi_features @ audio_features.t()
    
    return similarity_matrix

# Example usage:
# midi_features and audio_features should be the outputs from your model for a batch of paired samples
# accuracy = calculate_accuracy(midi_features, audio_features)
# print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
def train(model, dataloader, criterion, optimizer, device):
    model.train()
    running_losses = []
    running_loss = 0.0
    total_loss = 0.0
    total_accuracy = 0.0
    for batch_idx, (midi_data, audio_data) in enumerate(dataloader):
    # for batch in dataloader:
    #     midi_data, audio_data = batch  
        midi_data, audio_data = midi_data.to(device), audio_data.to(device)
        
        optimizer.zero_grad()
        
        # Forward pass
        midi_features, audio_features = model(midi_data, audio_data)
        loss = criterion(midi_features, audio_features)
        
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        running_loss += loss.item()
        # Calculate accuracy
        accuracy = calculate_accuracy(midi_features, audio_features)
        total_accuracy += accuracy
        
         # 打印Mini-batch训练信息
        if batch_idx % 10 == 9:  # 每 10 个 mini-batch 打印一次
            print('[Epoch %d, Batch %4d] Loss: %.4f' %
                  (epoch + 1, batch_idx + 1, running_loss / 10))
            running_losses.append(running_loss / 10)
            running_loss = 0.0  
            
    average_loss = total_loss / len(dataloader)
    average_accuracy = total_accuracy / len(dataloader)
    return average_loss, average_accuracy, running_losses #返回存储minibatch训练loss数据防止服务器中断训练数据丢失
```

```python
def validate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0.0
    total_accuracy = 0.0
    with torch.no_grad():
        for batch in dataloader:
            midi_data, audio_data = batch  
            midi_data, audio_data = midi_data.to(device), audio_data.to(device)
            
            # Forward pass
            midi_features, audio_features = model(midi_data, audio_data)
            loss = criterion(midi_features, audio_features)
            
            total_loss += loss.item()
            
            # Calculate accuracy
            accuracy = calculate_accuracy(midi_features, audio_features)
            total_accuracy += accuracy
            
    average_loss = total_loss / len(dataloader)
    average_accuracy = total_accuracy / len(dataloader)
    return average_loss, average_accuracy
```

```python
def save_list(data_list, filename):
    """
    将列表保存到文本文件中
    :param data_list: 要保存的列表
    :param filename: 文件名
    """
    with open(filename, 'w') as file:
        for item in data_list:
            file.write(str(item) + '\n')
```

```python
# Initialize lists to store training and validation losses
running_losses = []
train_losses = []
val_losses = []
train_accuracyes = []
val_accuracyes = []

# 训练循环
print("Start Training:")

for epoch in range(epochs):
    train_loss, train_accuracy, running_loss_list = train(model, train_loader, loss_fn, optimizer, device)
    val_loss, val_accuracy = validate(model, val_loader, loss_fn, device)

    # # Step the scheduler if using one
    # if scheduler:
    #     scheduler.step()

    print(f"Epoch {epoch+1}/{epochs}")
    print(f"Train Loss: {train_loss:.4f}")
    print(f"Train Accuracy: {train_accuracy * 100:.2f}%") #
    print(f"Validation Loss: {val_loss:.4f}")
    print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")
    
    #将训练数据存储
    running_losses += running_loss_list #将一个列表中的所有元素添加到另一个列表的末尾
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accuracyes.append(train_accuracy)
    val_accuracyes.append(val_accuracy)

    # Save model checkpoint if needed
    if val_accuracy > 0.9:
        torch.save(model.state_dict(), f"CELmodel_epoch_{epoch+1}.pth")
        save_list(running_losses, f'./result/CEL/running_losses_{epoch+1}.txt')
        save_list(train_losses, f'./result/CEL/train_losses_{epoch+1}.txt')
        save_list(val_losses, f'./result/CEL/val_losses_{epoch+1}.txt')
        save_list(train_accuracyes, f'./result/CEL/train_accuracyes_{epoch+1}.txt')
        save_list(val_accuracyes, f'./result/CEL/val_accuracyes_{epoch+1}.txt')
        
print("Finish Training!")
torch.save(model, 'CELmodel-fulldata.pth')
```

```python
# Define the Test Function
def test(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0.0
    total_accuracy = 0.0
    with torch.no_grad():
        for batch in dataloader:
            midi_data, audio_data = batch 
            midi_data, audio_data = midi_data.to(device), audio_data.to(device)
            
            # Forward pass
            midi_features, audio_features = model(midi_data, audio_data)
            loss = criterion(midi_features, audio_features)
            
            total_loss += loss.item()
            
            # Calculate accuracy
            accuracy = calculate_accuracy(midi_features, audio_features)
            total_accuracy += accuracy
            
    average_loss = total_loss / len(dataloader)
    average_accuracy = total_accuracy / len(dataloader)
    return average_loss, average_accuracy
```

```python
def extract_features(model, dataloader, device, is_midi=True):
    model.eval()
    all_features = []
    paths = []
    with torch.no_grad():
        for data, path in dataloader:
            data = data.to(device)
            if is_midi:
                features, _ = model(data, data)
            else:
                _, features = model(data, data)
            all_features.append(features) #.cpu())
            paths.extend(path)
    return torch.cat(all_features, dim=0), paths

# Extract MIDI library features
midi_library_features, midi_paths = extract_features(model, midi_library_loader, device, is_midi=True)
```

```python
def search_top_k(model, audio_query_loader, midi_library_features, midi_paths, logit_scale, k=5):
    model.eval()
    results = []
    with torch.no_grad():
        for audio_data, audio_path in audio_query_loader:
            audio_data = audio_data.to(device)
            _, audio_features = model(audio_data, audio_data)
            
            # Compute similarity scores
            similarity_scores = torch.matmul(audio_features, midi_library_features.t()) * logit_scale
            
            # Get top-k indices
            top_k_indices = torch.topk(similarity_scores, k, dim=1).indices.squeeze()
            
            # Get top-k paths
            top_k_paths = [midi_paths[i] for i in top_k_indices]
            results.append((audio_path, top_k_paths))
    return results
```

```python
class MidiSegmentDataset(Dataset):
    def __init__(self, midi_files, segment_length, overlap):
        self.midi_files = midi_files
        self.segment_length = segment_length
        self.overlap = overlap
        self.segments = []
        self.segment_paths = []
        self._create_segments() # This will call the method automatically during initialization
                
    def _create_segments(self):
        for midi_file in self.midi_files:
            midi_data = pretty_midi.PrettyMIDI(midi_file)
            midi_matrix = midi_data.get_piano_roll(fs=50) #[:128]  # Convert MIDI to piano roll and select first 128 pitches (time along columns)
            step = int(self.segment_length * (1 - self.overlap))
            for start in range(0, midi_matrix.shape[1] - self.segment_length + 1, step):
                segment = midi_matrix[:, start:start + self.segment_length]
                #normalize the data!
                midi_min, midi_max = np.min(segment), np.max(segment)
                if midi_max != midi_min:  # 避免除以零
                    segment = (segment - midi_min) / (midi_max - midi_min)
                else:
                    segment = np.zeros_like(segment)    
            
                self.segments.append(segment)
                self.segment_paths.append(midi_file)

    def __len__(self):
        return len(self.segments)

    def __getitem__(self, idx):
        segment = self.segments[idx]
        segment_path = self.segment_paths[idx]
        segment_tensor = torch.tensor(segment, dtype=torch.float32).unsqueeze(0)
        return segment_tensor, segment_path
```

```python
class AudioSegmentDataset(Dataset):
    def __init__(self, audio_file, segment_length, overlap):
        self.audio_file = audio_file # 单个文件名
        self.segment_length = segment_length
        self.overlap = overlap
        self.segments = []
        self.segment_paths = [] #就是audio_file,因为只让单个
        self._create_segments()

    def _create_segments(self):
        # for audio_file in self.audio_files:
        audio_data, sr = librosa.load(audio_file, sr=22050)
        # audio_matrix = librosa.feature.mfcc(y=audio_data, sr=sr, n_mfcc=20)
        mel_spectrogram = librosa.feature.melspectrogram(y=audio_data, sr=sr, n_fft=2048, hop_length=441, n_mels=128)
        audio_matrix = librosa.power_to_db(mel_spectrogram, ref=np.max)

        step = int(self.segment_length * (1 - self.overlap))
        for start in range(0, audio_matrix.shape[1] - self.segment_length + 1, step):
            segment = audio_matrix[:, start:start + self.segment_length]
            #normalize the data!
            audio_min, audio_max = np.min(segment), np.max(segment)
            if audio_max != audio_min:  # 避免除以零
                segment = (segment - audio_min) / (audio_max - audio_min)
            else:
                segment = np.zeros_like(segment)

            self.segments.append(segment)
            self.segment_paths.append(audio_file)

    def __len__(self):
        return len(self.segments)

    def __getitem__(self, idx):
        segment = self.segments[idx]
        segment_path = self.segment_paths[idx]
        segment_tensor = torch.tensor(segment, dtype=torch.float32).unsqueeze(0)
        return segment_tensor , segment_path
```

```python
def search_top_k_full_file(model, audio_dataloader, midi_library_features, midi_paths, segment_length, overlap, device, k):
#     audio_segments = preprocess_audio(audio_file, segment_length, overlap)
#     audio_features = extract_features_from_segments(model, audio_segments, device)
    
    audio_features , _ = extract_features_from_dataloader(model, audio_dataloader, device, is_midi=False) 
    # print("audio_features.size[0],num_segs: ", audio_features.shape[0])
    #audio_features is a tensor [num_segs, feature_dim]
    # print("midi_library_features'shape, [num_all_segs, fea-dim]",midi_library_features.shape)
    
    initial_logit_scale = np.log(1 / 0.07)
    logit_scale = torch.tensor(initial_logit_scale).exp().to(device)
    
    # Compute similarity scores for each audio segment
    similarity_scores = torch.matmul(audio_features.to(device), midi_library_features.to(device).t()) * logit_scale

    # Get top-k indices for each audio segment
    top_k_indices = torch.topk(similarity_scores, k, dim=1).indices.cpu().numpy()

    # Aggregate results to find the most frequently matched MIDI files
    midi_file_votes = {}
    for indices in top_k_indices:
        for idx in indices:
            midi_file = midi_paths[idx]
            if midi_file in midi_file_votes:
                midi_file_votes[midi_file] += 1
            else:
                midi_file_votes[midi_file] = 1
                
    # midi_file_votes = {}
    # for indices in top_k_indices:
    #     seen_segments = set()  # To track seen segments and avoid double counting
    #     for idx in indices:
    #         midi_file = midi_paths[idx]
    #         if idx not in seen_segments:
    #             seen_segments.add(idx)
    #             if midi_file in midi_file_votes:
    #                 midi_file_votes[midi_file] += 1
    #             else:
    #                 midi_file_votes[midi_file] = 1

    # Get the MIDI file with the most votes
    sorted_votes = sorted(midi_file_votes.items(), key=lambda item: item[1], reverse=True)
    return sorted_votes[:k]
```

```python
def search_top_k_full_file_batched(model, audio_dataloader, midi_library_features, midi_paths, segment_length, overlap, batch_size, device, k=5):
    #把k调大会不会效果好一点，每次只存前五是不是太少，不够更新轮换出准确的结果
    # audio_segments = preprocess_audio(audio_file, segment_length, overlap) #原来传入参数audio_dataloader位置传的是audio_file
    # audio_features = extract_features_from_segments(model, audio_segments, device).to(device)
    
    #the batch_size in param is for midi batch
    
    audio_features , _ = extract_features_from_dataloader(model, audio_dataloader, device, is_midi=False) 
    print("audio_features.size[0],num_segs: ", audio_features.shape[0])
    #audio_features is a tensor [num_segs, feature_dim]
    
    audio_features = audio_features.to(device)
    
    num_batches = (len(midi_library_features) + batch_size - 1) // batch_size
    
    # Initialize global top-k results for each audio segment
    global_top_k_indices = torch.full((len(audio_features), k), -1, dtype=torch.long, device=device)
    global_top_k_scores = torch.full((len(audio_features), k), float('-inf'), device=device)
    
    for i in range(num_batches):
        batch_start = i * batch_size
        batch_end = min((i + 1) * batch_size, len(midi_library_features))
        batch_features = midi_library_features[batch_start:batch_end].to(device)
        
        initial_logit_scale = np.log(1 / 0.07)
        logit_scale = torch.tensor(initial_logit_scale).exp().to(device)
        
        # Compute similarity scores for the current batch
        similarity_scores = torch.matmul(audio_features, batch_features.t()) * logit_scale
        
        # Update global top-k results
        for j in range(similarity_scores.shape[0]):
            scores = similarity_scores[j]
            combined_scores = torch.cat((global_top_k_scores[j], scores))
            combined_indices = torch.cat((global_top_k_indices[j], torch.arange(batch_start, batch_end, device=device)))
            topk_scores, topk_indices = torch.topk(combined_scores, k)
            global_top_k_scores[j] = topk_scores
            global_top_k_indices[j] = combined_indices[topk_indices]
    
    # Aggregate votes to determine the most frequently matched MIDI files
    midi_file_votes = {}
    for indices in global_top_k_indices:
        for idx in indices:
            midi_file = midi_paths[idx]
            if midi_file in midi_file_votes:
                midi_file_votes[midi_file] += 1
            else:
                midi_file_votes[midi_file] = 1
    
    # Sort the results based on votes
    sorted_votes = sorted(midi_file_votes.items(), key=lambda item: item[1], reverse=True)
    return sorted_votes[:k]
```
